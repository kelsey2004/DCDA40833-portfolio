## Aggregate Daily Phone Hours by Age

### Subtask:
Group the dataset by 'Age' and calculate the sum of 'Daily_Phone_Hours' for each age group, then display the aggregated data.


**Reasoning**:
To aggregate the daily phone hours by age, I will group the DataFrame by the 'Age' column and then sum the 'Daily_Phone_Hours' for each age group.



```python
daily_phone_hours_by_age = df.groupby('Age')['Daily_Phone_Hours'].sum() # I understand the logic.It is grouping age by Daily Phone Hours becasue it needs to calculate the total for each age.
print("Total Daily Phone Hours by Age:\n")
print(daily_phone_hours_by_age.head())
```

## Visualize Total Daily Phone Hours by Age

### Subtask:
Create a bar chart with 'Age' on the X-axis and the aggregated 'Daily_Phone_Hours' on the Y-axis. Ensure the plot has appropriate labels and a legend.


**Reasoning**:
To visualize the total daily phone hours by age, I will create a bar chart using the aggregated data, ensuring it has appropriate labels and a title for clarity.



```python
import matplotlib.pyplot as plt

# Set the figure size for better readability
plt.figure(figsize=(15, 7))

# Create the bar chart
plt.bar(daily_phone_hours_by_age.index, daily_phone_hours_by_age.values) # I understand it it taking the age and phone hours in order to create the bar chart.

# Add labels and title
plt.xlabel('Age') # Adding the string "Age" in order to label the X axis
plt.ylabel('Total Daily Phone Hours') # Adding the string "Total Daily Phone Hours" in order to label the Y axis
plt.title('Total Daily Phone Hours by Age') #Adding the string "Total Daily Phone Hours by Age" in order to give the chart at title.

# Rotate x-axis labels if there are too many to prevent overlap
# To make every single age value on the x-axis visible:
plt.xticks(daily_phone_hours_by_age.index, rotation=45, ha='right')

# Add a legend if multiple series were plotted (not applicable here, but good practice)
# For a single bar chart, a legend is usually not needed, but can be simulated if required for a specific purpose

# Improve layout and display the plot
plt.tight_layout()
plt.show() #Displaying the bar chart
```
